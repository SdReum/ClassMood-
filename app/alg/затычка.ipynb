```python
import cv2 
import pandas as pd
from random import uniform, random
```

```python
class BespoleznayaHuinya:
    def __init__(self):
        pass

    def nicho_ne_delait(self, path):
        cap = cv2.VideoCapture(path)
        n = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
        df = pd.DataFrame({
            'frame': list(range(0, n, 5)),
            'yaw pitch roll': [[(uniform(-90, 90), uniform(-90, 90), uniform(-90, 90)) for _ in range(30)] for _ in range(int(n // 5))]
        })

        return df

    def nicho_ne_delait_to_interest(self, path):
        df = self.nicho_ne_delait(path)
        res = pd.DataFrame({
            'frame': df['frame'],
            'interest': [random() for _ in range(len(df))]
        })

        return res
```
